```python
import numpy as np
import pandas as pd
```

```python
emp = pd.DataFrame({'Emp_ID':[101,102,103,104,105],
                   'Ename':['Sana','Kane','Riya','Peter','Sandy'],
                   'Dept_ID':[11,22,22,44,66]})
emp
```

```python
dept = pd.DataFrame({"Dept_ID":[11,22,33,44],
                    "Dept_name":['Admin',"HR","Software Dev.",'Sales'],
                    'Location':['D0','D1','D1','D3']})
dept
```

```python
pd.merge(emp,dept, how='inner',on="Dept_ID")
```

```python
pd.merge(emp,dept, how='left',on="Dept_ID")
```

```python
pd.merge(emp,dept, how='outer',on="Dept_ID")
```

```python
dept2 =pd.DataFrame({'D_ID':[11,22,33,44],
                    'Dname':['Admin','Software Devlopment','HR','Sales'],
                    'Location':['DO','D1','D2','D3']})
dept2
```

```python
pd.merge(emp, dept2, left_on ='Dept_ID',right_on='D_ID')
```

```python
emp3 =pd.DataFrame({'Emp_ID':[111,102,102,104,105],
                   'Ename':['Sana','Kane','Riya','Peter','Sandy'],
                   'Skills':['SQL','Spark','Hadoop','Docker','Tableau']})
emp3
```

```python
pd.merge(emp, emp3,on=['Emp_ID','Ename'])
```

```python
pd.merge(emp, emp3, on='Emp_ID')
```

```python
pd.merge(emp, emp3, on='Emp_ID', suffixes=('_emp','_emp3'))
```

```python
dept4 = pd.DataFrame({'Dname':["Admin",'Software Devloper',"HR","Sales"],
                     'Location':['D0','D1','D1','D2'],
                     },
                    index = [11,22,33,44])
dept4
```

```python
pd.merge(emp, dept4,left_on="Dept_ID", right_index= True)
```

```python
emp4 =pd.DataFrame({'Emp_ID':[101,102,102,104,105],
                   'Ename':['Sana','Kane','Riya','Peter','Sandy'],
                   },
                   index = [11,22,33,44,66])
emp4
```

```python
pd.merge(emp4,dept4, how = 'outer', left_index = True, right_index = True)
```
