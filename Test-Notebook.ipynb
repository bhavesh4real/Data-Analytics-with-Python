```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family'] = 'STIXNonUnicode'
```

```python
data = {'Team': ['India', 'Australia', 'Pakistan', 'South Africa', 'England'],
        'Matches': [100, 90, 80, 70, 60],
        'Runs': [5000, 4500, 4000, 3500, 3000],
        'Average_Runs': [50, 45, 40, 35, 30],
        'Average_SR': [80, 75, 70, 65, 60]}
df = pd.DataFrame(data)
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bars
bars = ax.bar(df.index, df.Runs, color=['#0072C6', '#FF9933', '#006600', '#007A5E', '#CE1126'])

# Add national flag emojis to the bars
ax.text(0.05, df.Runs[0] + 200, u'\U0001F1EE\U0001F1F3', fontsize=20) # India
ax.text(1.05, df.Runs[1] + 200, u'\U0001F1E6\U0001F1FA', fontsize=20) # Australia
ax.text(2.05, df.Runs[2] + 200, u'\U0001F1F5\U0001F1F0', fontsize=20) # Pakistan
ax.text(3.05, df.Runs[3] + 200, u'\U0001F1FF\U0001F1E6', fontsize=20) # South Africa
ax.text(4.05, df.Runs[4] + 200, u'\U0001F3F4\U000E0067\U000E0062\U000E0065\U000E006E\U000E0067\U000E007F', fontsize=20) # England

# Set the x-tick labels and the title
ax.set_xticks(df.index)
ax.set_xticklabels(df.Team)
ax.set_title('Runs Scored by Teams', fontsize=16)
```
