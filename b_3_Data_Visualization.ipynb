* Subpolts of containing different function's graph.

```python
x = np.arange(0,3 * np.pi, 0.1)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.plot(np.random.randn(8))
ax2.plot(np.sin(x))
ax3.plot(np.cos(x))
```

```python
fig, axes = plt.subplots(2,2, sharex = True, sharey= True)
for i in range(2):
    for j in range(2):
        axes[i,j].plot(np.random.randn(10))

plt.subplots_adjust(wspace = 0, hspace = 0)
```

```python
fig, axes = plt.subplots(2,2, sharex = True, sharey= True)
for i in range(2):
    for j in range(2):
        axes[i,j].plot(np.random.randn(10))

plt.subplots_adjust(wspace = 0.2, hspace = 0.2)
```

### More basic plotting features.
1. Line styling
2. Changing colors
3. Using different markers
4. Adding title, axis lables, and legend.


```python
data = np.random.randn(30)
data
```

```python
plt.plot(data,linestyle = "--", color = 'b' ,marker = 'o')
```

```python
plt.plot(data, color = 'g', linestyle = '-.', marker = 'd' )
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(data)
ax.set_xticks([0,10,20,30])
ax.set_xticklabels(['Zero','Ten','Twenty','Thirty'],
                  rotation = 20, fontsize = 'small')

ax.set_title('THIS IS THE PLOT TITLE AS: TEST')


ax.set_xlabel('Stage')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.sin(x),'b--', label = 'Sine Wave')
ax.plot(np.cos(x),'g-.', label = "Cosine Wave")
ax.legend()
plt.savefig('superimpose_sin_cos.png')
```
