```python
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
df = pd.read_csv("company_sales_data.csv")
```

```python
df.head()
```

####  1: Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties: –

    X label name = Month Number
    Y label name = Total profit
The line plot graph should look like this.

```python
month_num_x = df['month_number']
total_profit_y = df['total_profit']

#-adjusting appropriate figure size
plt.figure(figsize=(5.5,3.5), dpi=80)

plt.plot(month_num_x,total_profit_y)

plt.xlabel('Month number')
plt.ylabel('Profit in dollar')

#modifying ticks on both the axis
plt.xticks(month_num_x)
plt.yticks([100000,200000,300000,400000,500000])

plt.title('Company profit per month')
```

####  2: Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties: –

    Line Style dotted and Line color should be red
    Show legend at the lower right location.
    X label name = Month Number
    Y label name = Profit
    Add a circle marker.
    Line marker color as read
    Line width should be 3

The line plot graph should look like this.

```python
month_num_x = df['month_number']
total_profit_y = df['total_profit']

#-adjusting appropriate figure size
plt.figure(figsize=(5.5,3.5), dpi=80)

plt.plot(month_num_x,total_profit_y,'ro--',mfc='k',linewidth = 3)

plt.xlabel('Month number')
plt.ylabel('Profit in dollar')

#modifying ticks on both the axis
plt.xticks(month_num_x)
plt.yticks([100000,200000,300000,400000,500000])

plt.title('Company profit per month')
plt.legend(['Profit data of last year'],loc='lower right')
```

#### 3: Read all product sales data and show it  using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

The graph should look like this.

```python
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-bright')

plt.plot(month_num_x,facecream, marker = 'o',linewidth = 2.5, label ='Face Cream')
plt.plot(month_num_x,facewash, marker = 'o',linewidth = 2.5, label ='Face Wash')
plt.plot(month_num_x,toothpaste, marker = 'o',linewidth = 2.5, label ='ToothPaste')
plt.plot(month_num_x,bathingsoap, marker = 'o',linewidth = 2.5, label ='Bathingsoap')
plt.plot(month_num_x,shampoo, marker = 'o',linewidth = 2.5, label ='Shampoo ')
plt.plot(month_num_x,moisturizer, marker = 'o',linewidth = 2.5, label ='Moisturizer')

plt.xlabel('Month number')
plt.ylabel('Sales units in numbers')

#modifying ticks on both the axis
plt.xticks(month_num_x)

plt.title('Sales Data')
plt.legend(loc='best')
```

#### 4: Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. grid line style should “–“.

The scatter plot should look like this.

```python
plt.scatter(month_num_x,toothpaste,cmap= 'viridis',edgecolor='k')

plt.title('Tooth paste Sales data.')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')

plt.legend(['Tooth paste Sales data'])
plt.grid(True,ls ='--')
```

#### 5: Read face cream and facewash product sales data and show it using the bar chart

Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

```python
plt.figure(figsize=(7,4), dpi=80)
wd = 0.25

plt.bar(month_num_x,facecream,width =wd,color='#1E90FF',label='Face Cream sales data')
plt.bar(month_num_x + wd,facewash,width =wd,color='#FF7F50',label='Face Cream sales data')


plt.title('FaceWash and FaceCream Sales data.')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')

plt.xticks(month_num_x)

plt.legend(loc='upper left')
plt.grid(True,ls ='--')
```

#### 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
plt.figure(figsize=(5.5,3.3))
plt.bar(month_num_x,bathingsoap,color='#1E90FF')

plt.xticks(month_num_x)

plt.title('bathingsoap sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

plt.grid(True,ls='--')
#plt.savefig('exercise_plot.png')
```

#### 7: Read the total profit of each month and show it using the histogram to see most common profit ranges

The histogram should look like this.

```python
plt.figure(figsize=(5.4,3.3))
bin_tags =[150000,175000,200000,225000,250000,300000,350000]
plt.hist(total_profit_y,bins=bin_tags,color = '#1E90FF',edgecolor='k',label='Profit Data')

plt.title('Profit Data')
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual profit in dollar')

plt.xticks(rotation = 30)

plt.legend(loc='upper left',fontsize=8,framealpha =0.8)
```

#### 8: Calculate total sale data for last year for each product and show it using a Pie chart

Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.

```python
slices =[df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
label_tags =['FaceCream','FaceWash','ToothPaste','Bathing Soap','Shampoo','Moisturizer']
color_tags =['#4682B4','#FFA500','#32CD32','#B22222','#9370DB','#D2691E']

explode_tuple =(0,0.1,0,0,0,0)

plt.figure(figsize=(4,5))
plt.pie(slices,labels=label_tags,colors=color_tags,autopct ='%1.0f%%',explode = explode_tuple ,wedgeprops={'edgecolor':'k'},shadow = True)
plt.legend(loc=(1.2,0))

plt.title('Previous Year\'s Total Sales of each Product.')
```

#### 9: Read Bathing soap facewash of all months and display it using the Subplot

The Subplot should look like this.

```python
fig,(ax1,ax2) = plt.subplots(2,1, sharex =True)

ax1.plot(month_num_x,bathingsoap,'ko-',linewidth =2.4)

ax2.plot(month_num_x,facewash,'ro-',linewidth =2.4)

ax1.set_title('Bathing Soap Sales Data')
ax2.set_title('Face Wash Sales Data')

plt.xticks(month_num_x)

plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')
```
