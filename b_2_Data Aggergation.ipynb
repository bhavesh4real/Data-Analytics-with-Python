### Data Aggegration in Pandas
Some data aggregation functions and methods are used following in the notebook.

```python
import numpy as np
import pandas as pd
```

#### Some Basic *Data Aggegration* Techniques

```python
#Creting a dataframe df
df = pd.DataFrame({
    'key1':['a','a','b','b','a'],
    'key2':['one','two','one','two','one'],
           'data1':np.random.randn(5),
           'data2':np.random.randn(5)
})

df
```

```python
#creating a group object 'grouped' group by column 'key1'
grouped = df.groupby('key1')
grouped
```

```python
grouped['data1'].mean()
```

```python
means = df.groupby(['key1','key2'])['data1'].mean()
means
```

```python
means.unstack()
```

```python
df.groupby('key1').max()
```

```python
df.groupby(['key1','key2'])['data1'].mean()
```

```python
df.groupby(['key1','key2']).size()
```

#### Iteration on *Group By* can be performed as follows:

```python
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
for (k1,k2), group in df.groupby(['key1','key2']):
    print((k1,k2))
    print(group)
```

```python
df
```

```python
grouped = df.groupby(['key1'])
grouped['data1'].max()
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
```

```python
grouped.agg(peak_to_peak)
```

### Working with **Tips** *Dataset*

```python
bill_data = pd.read_csv("tips.csv")
bill_data.tail()
```

```python
bill_data.info()
```

#### This is not a huge data but it still contains 244 entries for 8 column titles.

*We can attempt to make all the rows visible by changing some pandas settings.*

```python
pd.set_option('display.max_rows',244)
```

```python
bill_data
```

```python
#adding new column tip percentage
bill_data['tip_pct'] = bill_data['tip']/bill_data['total_bill']
bill_data.head()
```

```python
#Creating group object by each day in a week and the smoker type.
grouped = bill_data.groupby(['day', 'smoker'])
```

```python
gender = grouped['gender']
```

```python
gender.value_counts()
```

* if we are intrested to know average tip per cent paid by smokers and non-smoker on some days 

```python
grouped_pct = grouped['tip_pct']
grouped_pct.mean()
```

We can see that, the smokers are tipping 2% more than non-smokers on Sundays and Fridays.

Whereas, on Saturdays & Thrusdays the avgerage tips by smokers & non-smokers are nearly the same or differ by just a percent.

```python
#peak to perk val, mean & st. deviation in tip percent 
grouped_pct.agg(['mean','std',peak_to_peak])
```

```python
#renaming our column titles for our convenience
grouped_pct.agg([('Mean_Tip', 'mean'),('Std_Tip', 'std')]) 
```

```python
#user defined func top returns sorted values by 'tip_pct' in the ascending order.
#the default parameter is set to 5, so by default it will return 5 rows.
def top(df, n=5, column = 'tip_pct'):
    return df.sort_values(by=column)[-n:]

top(bill_data)
```

```python
#check top 7 tip_pcts in tips dataset
top(bill_data, 7)
```

```python
bill_data.groupby('smoker').apply(top)
```

```python
bill_data.groupby(['smoker','day']).apply(top, n=1, column = 'total_bill')  
```

```python
bill_data.groupby(['smoker','day']).apply(top, n=2, column = 'total_bill')  #for n = 2 top two rows are returned
```

```python
bill_data.groupby('smoker',group_keys = False ).apply(top)
```

```python
cities = ['New Delhi', 'Jaipur','Patna','Indore','Nagpur','Hyderabad','Chennai','Kochi']
group_key = ["North"]*4 + ['South']*4
group_key
```

```python
data = pd.Series(np.random.randn(8), index = cities)
data[['Jaipur','Indore','Chennai']] = np.nan
data
```

```python
data.groupby(group_key).mean()
```

```python
fill_mean = lambda g: g.fillna(g.mean())
data.groupby(group_key).apply(fill_mean)
```

```python
fill_values = {'North':0.5, 'South': -1}
fill_func = lambda g: g.fillna(fill_values[g.name])
data.groupby(group_key).apply(fill_func)
```
